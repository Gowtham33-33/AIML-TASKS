```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Netflix_Dataset.csv")
print(data)
```

```python
data.info()
```

```python
print(type(data))
print(data.shape)
print(data.size)
```

```python
data1 = data.drop(["show_id"],axis =1)
data1
```

```python
data1['release_year']=pd.to_numeric(data['release_year'],errors='coerce')
data1.info()
```

```python
data1[data1.duplicated(keep=False)]
```

```python
data1.drop_duplicates(keep='first',inplace = True)
data1
```

```python
data1.rename({'duration_minutes': 'duration'},axis =1,inplace = True)
data1
```

```python
data1.isnull().sum()
```

```python
cols = data1.columns
colors = ['black','red']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
median_duration = data1["duration"].median()
mean_duration = data1["duration"].mean()
print("Median of duration: ",median_duration)
print("Mean of duration: ",mean_duration)
```

```python
data1['duration'] = data1['duration'].fillna(median_duration)
data1.isnull().sum()
```

```python
mean_duration_seasons = data1["duration_seasons"].mean()
print("Mean of duration_seasons: ",mean_duration_seasons)
```

```python
data1['duration_seasons'] = data1['duration_seasons'].fillna(mean_duration_seasons)
data1.isnull().sum()
```

```python
print(data1["type"].value_counts())
mode_type = data1["type"].mode()[0]
print(mode_type)
```

```python
data1["type"] = data1["type"].fillna(mode_type)
data1.isnull().sum()
```

```python
print(data1["release_year"].value_counts())
mode_type = data1["release_year"].mode()[0]
print(mode_type)
```

```python
data1["release_year"] = data1["release_year"].fillna(mode_type)
data1.isnull().sum()
```

```python
fig, axes = plt.subplots(2, 1,figsize=(8,6), gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["duration"],ax=axes[0],color='red',width=0.5,orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("duration Levels")
sns.histplot(data1["duration"],kde=True, ax=axes[1],color='purple',bins=30)
axes[1].set_title("Histogram with kde")
axes[1].set_xlabel("duration levels")
axes[1].set_ylabel("frequency")
plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(2, 1,figsize=(8,6), gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["duration_seasons"],ax=axes[0],color='yellow',width=0.5,orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("duration_seasons Levels")
sns.histplot(data1["duration_seasons"],kde=True, ax=axes[1],color='blue',bins=30)
axes[1].set_title("Histogram with kde")
axes[1].set_xlabel("duration_seasons levels")
axes[1].set_ylabel("frequency")
plt.tight_layout()
plt.show()
```

```python
data1["duration"].describe()
```

Observations
- It is observed that only two outliers are identified by sd
- In the box plot method we can find more outliers 
- This is because the assumption of normality is not satisfied in this column

```python
mu=data1["duration"].describe()[1]
sigma = data1["duration"].describe()[2]
for x in data1["duration"]:
    if ((x < (mu - 3*sigma)) or (x > (mu + 3*sigma))):
        print(x)
```

```python
import scipy.stats as stats
plt.figure(figsize=(8, 6))
stats.probplot(data1["duration"],dist="norm",plot=plt)
plt.title("Q-Q Plot for Outliers Detection",fontsize=14)
plt.xlabel("Theretical Quantiles",fontsize=12)
```

### Observations in Q-Q plot
- The data does not follow normal distribution as the data points are deviating significally away from the red line
- The data shows a right-skweed distribution and possible of outlier
